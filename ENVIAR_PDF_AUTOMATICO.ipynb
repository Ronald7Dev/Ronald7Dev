```python
import requests
from google.colab import files

# URL do webhook no n8n (substitua pela sua)
url = "https://ronald7dev.app.n8n.cloud/webhook-test/receive-pdf"

# Fazer upload do arquivo manualmente no Google Colab
print("Selecione o arquivo PDF para enviar ao n8n:")
uploaded = files.upload()  # Irá abrir uma janela para selecionar o PDF

# Pega o nome do arquivo carregado
arquivo_pdf = list(uploaded.keys())[0]

# Enviar o PDF via POST para o n8n
with open(arquivo_pdf, "rb") as file:
    files = {"file": file}
    response = requests.post(url, files=files)

# Exibir resposta do servidor
print("Status Code:", response.status_code)
print("Resposta:", response.text)
```
